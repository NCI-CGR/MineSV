# Introduction

This study is to benchmark the MoCCA SV results using bedtools and jupyter notebook.  To start with the same envrionmental setting as Bari, I have disabled the local conda setting:
+ conda deactivate
+ module load tmux python3 bedtools zlib

We are going to process the data in the following steps: 
+ Convert the truth VCF to bed format.
+ Select MoCCA predictions included in the regions specified by GIAB (for GIAB benchmark only).
+ Padd Inserts with N bp (N=500, by default) on both the truth bed file and the MoCCA bed file.
+ Find the overlapped btw truth and MoCCA bed files with over 70% reciporal overlap.
+ Collect the summary of the result. 
    
The process will be repeated for each SV callers and the combination of the callers. 
[:memo:] We will benchmark intra-chromosomal SVs only as there are not true intra-chromosomal SV defined in GIAB or the synthetic data set.

---
# Task specification
Overall goal: Benchmark germline MoCCA-SV results against GiaB HG002/003/004, with the aim of establishing the specificity/sensitivity ratios of various filtering strategies
Initial strategy:
- Subset truth and pipeline output to the GiaB Tier 1 regions of interest (e.g. using bedtools intersect with the -u option)
- Subdivide pipeline output into SVs called by one, two, or three callers
- Isolate potential insertions in the truth and pipeline output by using a length cutoff (e.g. less than X bp between start and end), and pad (e.g. by 1000 bases up- and down-stream), then recombine
- Intersect the truth and pipeline output to get the TP SVs (e.g. bedtools intersect with -u, -f, and -r options).  Note that the -f value selected here will dictate the actual window to compare insertions, e.g. -f 0.9 and 500 bp up/down-stream padding means that insertions will be counted as concordant if they are within (500 + 500) * 0.1 = 100 bases.
- Intersect as above, but with the -v flag instead of -u, to get the FN count
- Intersect as above, but write out the intersecting pipeline output instead of the truth data to get the FP SV calls.  Note that this data will need to be collapsed to give an accurate FP count; there is one row per caller-per SV as-is.
- The counts derived above can be used to calculate recall/precision/F1/etc.
- Note that this workflow covers intrachromosomal SVs only, as there are no interchromosomal SVs in the GiaB truth dataset
Subsequent strategies:
- Depending on the results of the first round, we anticipate at least 2-3 additional rounds of benchmarking, using different filtering strategies (e.g. a combination of caller concordance and other metrics, and/or adjustments of the reciprocal overlap required to determine caller concordance).  The choice of metrics in these strategies will be guided by the analysis of the initial strategy.
Deliverables:
- Jupyter notebook containing all commands necessary to perform the initial strategy
- The notebook should be written to be easily applied to other benchmarking projects, e.g. using variables to define paths and filenames that can be set once for each notebook run
- Similar Jupyter notebooks for each subsequent strategy tested
- All intermediate and final data files generated by the steps run in the notebooks
Future work:
- Additional projects may include selection of additional SV callers, implementation within the pipeline, and additional benchmarking to establish filtering recommendations

# Data analyses
## GIAB data set
### 1. Intersect MoCCA results with the included regions

```python
%cd /mnt/nfs/gigantor/ifs/DCEG/Projects/CoherentLogic/SV/mocca-bench/data
!ls -al
```

```python
%%bash
bedtools intersect -a intrachromosomal_SVs_AJ_2x250_son -b HG002_SVs_Tier1_noVDJorXorY_v0.6.2.bed -f 1.0000000 -u > mocca_giab_included.bed
wc -l mocca_giab_included.bed
```

```python
!head mocca_giab_included.bed
```

### 2. Padding the insertions 

```python
# deffine some constants in python 
N=500 # for padding intersion
min_ovp=0.7 # for minimum reciprocal overlap
max_ins_size=20 # assume that the insert size is no more than 20
```

```python
%%bash -s "$max_ins_size" "$N"
echo $1, $2
awk -v OFS='\t' -v var="$1" -v pad="$2" '{if($3-$2 <= var){$3=$3+pad;$2=$2-pad} print }' mocca_giab_included.bed > mocca_giab_padded.bed
```

```python
!diff mocca_giab_included.bed mocca_giab_padded.bed | head
```

```python
# giab insersion is different 
! grep INS my-tp-base.bed | head
```

```python
%%bash -s "$max_ins_size" "$N"
echo $1, $2
awk -v OFS='\t' -v var="$1" -v pad="$2" '{if($5 == "INS"){$3=$3+pad;$2=$2-pad} print }' my-tp-base.bed > truth_giab_padded.bed
```

```python
! diff my-tp-base.bed  truth_giab_padded.bed | head
```

### 3. Get the intersects

```python
%%bash -s "$min_ovp"

bedtools intersect -a truth_giab_padded.bed -b mocca_giab_padded.bed -wa -wb -f 0.7 -r > truth_giab_predicted.bed
```

```python
! head truth_giab_predicted.bed
```

### 4. Summary the results
[:memo:] We svaba, breakdancer and manta results in the order in the MoCCA-SV output.

"#CHROM	start	end	svaba	breakdancer	manta	caller_count"	

```python
%%bash --out tp_cnt

array=(10 11 12)
for i in "${array[@]}"
do
    awk -v "z=$i" '{if($z=="orig") print $4}' truth_giab_predicted.bed | sort -u |wc -l
done
#awk '{if($10=="orig") print $4}' truth_giab_predicted.bed | sort -u |wc -l
#echo 1
```

```python
#manually confirm
!awk '{if($10=="orig") print $4}' truth_giab_predicted.bed | sort -u | wc -l
!awk '{if($11=="orig") print $4}' truth_giab_predicted.bed | sort -u | wc -l
!awk '{if($12=="orig") print $4}' truth_giab_predicted.bed | sort -u | wc -l
```

```python
%%bash --out pred_cnt

array=(4 5 6)
for i in "${array[@]}"
do
    awk -v "z=$i" '{if($z=="orig") print}' mocca_giab_included.bed | wc -l
done
#awk '{if($10=="orig") print $4}' truth_giab_predicted.bed | sort -u |wc -l
#echo 1
```

```python
import pandas as pd

def get_nums_from_bash_std(lines):
    _n=list(filter(None, lines.split('\n')))
    rv = map(lambda x: int(x), _n)
    return rv

cc = !cat my-tp-base.bed |wc -l
# giab_germline_res = pd.DataFrame({"GIAB":int(cc[0]), "TP":map(lambda x: int(x), _tp)}, index=['svaba', 'breakdancer', 'manta'])

df = pd.DataFrame({"GIAB":int(cc[0]), "TP":get_nums_from_bash_std(tp_cnt), 
                   "PRED":get_nums_from_bash_std(pred_cnt)}, index=['svaba', 'breakdancer', 'manta']).assign(
                Sp=lambda df: df.TP/df.PRED, 
                Sn=lambda df: df.TP/df.GIAB)

df.index.name="Caller" 
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
g=sns.scatterplot(x="Sn", y="Sp", data=df.reset_index(), hue="Caller")
g.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)
plt.show()
```
